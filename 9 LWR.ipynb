```python
def rad_ker(x2,x1,tau):
    return np.exp(np.sum((x1-x2)**2,axis=1)/(-2*tau**2))
   
```

```python
# def loc_reg(x1,x2,y1,tau):
#     x1 = np.r_[np.ones(len(x1)),x1]
#     x2 = np.c_[1,x2]
#     xw = x1.T * rad_ker(x1,x2,tau)
#     beata = np.linalg.pinv(xw@x1)@xw@y1
#     return x2@beta

def local_regression(x0, x, y, tau):
    x0 = np.r_[1, x0]
    x = np.c_[np.ones(len(x)), x]
    xw = x.T * rad_ker(x0, x, tau)
    beta = np.linalg.pinv(xw @ x) @ xw @ y
    return x0 @ beta
```

```python
def plot_lr(tau):
    x2 = np.linspace(-5,5,300)
    y2 = [local_regression(x2,x,y,tau) for x2 in x2]
    plt.scatter(x2,y2)
    plt.plot(x2,y2)
    
plot_lr(0.02)

# def plot_lr(tau):
#     domain = np.linspace(-5, 5, num=300)
#     pred = [local_regression(x0, x, y, tau) for x0 in domain]
#     plt.scatter(x, y, alpha=0.3)
#     plt.plot(domain, pred, color="red")
#     return plt
```
