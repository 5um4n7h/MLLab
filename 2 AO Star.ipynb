```python
class Graph:
    
    def __init__(self,start,h,g):
        self.start = start
        self.h = h
        self.g = g
        self.parents = {}
        self.status = {}
        self.sG = {}
    
    
    def applyAO(self):
        self.AOStar(self.start,False)
    
    def getH(self,v):
        return self.h.get(v,0)
    
    def setH(self,v,val):
        self.h[v] = val
        
    def getStatus(self,v):
        return self.status.get(v,'')
    
    def setStatus(self,v,val):
        self.status[v] = val
        
    def getNbr(self,v):
        return self.g.get(v,'')
    
    def sol(self):
        for item in reversed(self.sG):
            print(item)
            
    def minCostNodes(self,v):
        
        minC = 0
        costtoChNodeDict = {}
        costtoChNodeDict[minC] = []
        flag = True
        
        
        for chNodeList in self.getNbr(v):
            cost = 0
            nodeList = []
            
            for node,w in chNodeList:
                nodeList.append(node)
                cost = cost+self.getH(v)+w
                
            if flag:
                minC = cost
                costtoChNodeDict[minC] = nodeList
                flag = False
                
            elif minC > cost:
                minC = cost
                costtoChNodeDict[minC] = nodeList
                
        return minC,costtoChNodeDict[minC]
    
    def AOStar(self,v,bT):
        
        print("H : ",self.h)
        print("Sol G : ",self.sG)
        print("P node : ",v)
        
        minC,chNodeList = self.minCostNodes(v)
        self.setH(v,minC)
        self.setStatus(v,len(chNodeList))
        
        if self.getStatus(v)!=-1:
            
            solved = True
            
            for node in chNodeList:
                self.parents[node] = v
                if self.getStatus(node)!=-1:
                    solved = False
                    
            if solved :
                self.setStatus(v,-1)
                self.sG[v] = chNodeList
                
            if v!=self.start:
                print(self.parents)
                self.AOStar(self.parents[v],True)
                
                
            if not bT:
                for node in chNodeList:
                    self.setStatus(node,0)
                    self.AOStar(node,False)
                    
                    
            
            
        
    
                
            
            
            
```

```python
h1 = {'A': 0, 'B': 6, 'C': 12, 'D': 10, 'E': 4, 'F': 4, 'G': 5, 'H': 7, 'I': 1, 'J': 1}
graph1 = {
    'A': [[('B', 1), ('C', 1)], [('D', 1)]],
    'B': [[('G', 1)], [('H', 1)]],
    'C': [[('J', 1)]],
    'D': [[('E', 1), ('F', 1)]],
    'G': [[('I', 1)]]
}

G1 = Graph('A',h1,graph1)
G1.applyAO()
G1.sol()
```
